```python
#import libraries
import pandas as pd

#read csv's and assign identities
#change /filepath/ to your corresponding file path and change filename (stats.csv) if necessary
FIR = pd.read_csv("/filepath/stats.csv")
GIR = pd.read_csv("/filepath/stats-2.csv")
Scrambling = pd.read_csv("/filepath/stats-3.csv")
Ovr_Putting_Avg = pd.read_csv("/filepath/stats-4.csv")
SG_Tee = pd.read_csv("/filepath/stats-5.csv")
SG_Approach = pd.read_csv("/filepath/stats-6.csv")
SG_ARG = pd.read_csv("/filepath/stats-7.csv")
SG_Putting = pd.read_csv("/filepath/stats-8.csv")
FedEx_Cup_Points = pd.read_csv("/filepath/stats-9.csv")
```

```python
#data cleaning: removing redundant columns
df_list = [FIR, GIR, Scrambling, Ovr_Putting_Avg, SG_Tee, SG_Approach, SG_ARG, SG_Putting] 
for i in df_list: 
    i.drop(columns = ['RANK', 'MOVEMENT'], inplace=True)
```

```python
#merging all statistics together 
merge_1 = pd.merge(FIR, GIR, on=['PLAYER_ID','PLAYER'], how='inner', suffixes = ('_FIR', '_GIR'))
merge_2 = pd.merge(merge_1, Scrambling, on=['PLAYER_ID','PLAYER'], how='inner', suffixes = ('_Scrambling', '_GIR'))
merge_3 = pd.merge(merge_2, Ovr_Putting_Avg, on=['PLAYER_ID','PLAYER'], how='inner')
merge_4 = pd.merge(merge_3, SG_Tee, on=['PLAYER_ID','PLAYER'], how='inner', suffixes = ('_Putting', '_SG_Tee'))
merge_5 = pd.merge(merge_4, SG_Approach, on=['PLAYER_ID','PLAYER', 'MEASURED ROUNDS'], how='inner', suffixes = ('_SG_Tee', '_SG_APP'))
merge_6 = pd.merge(merge_5, SG_ARG, on=['PLAYER_ID','PLAYER', 'MEASURED ROUNDS'], how='inner', suffixes = ('_SG_APP', '_SG_ARG'))
merge_7 = pd.merge(merge_6, SG_Putting, on=['PLAYER_ID','PLAYER', 'MEASURED ROUNDS'], how='inner', suffixes = ('_SG_ARG','_SG_Putting'))
merge_7
```

```python
#creating final working dataset merging the statistics to the FedEx Cup points
df = pd.merge(FedEx_Cup_Points, merge_7, on=['PLAYER_ID','PLAYER'], how = 'left')

#Data Cleaning:
#Removing Unnecessary column
df = df.drop(columns = 'MOVEMENT')
#renaming columns for clarity
df = df.rename(columns = {'%':'%_Scrambling', 'AVG':'AVG_SG_PUTTING'})

#turning % columns into floats rather than percents so that we can use them in calculations
df['%_FIR']=df['FAIRWAYS HIT']/df['POSSIBLE FAIRWAYS']
df['%_GIR']=df['GREENS HIT']/df['# HOLES']
df['%_Scrambling']=df['PAR OR BETTER']/df['MISSED GIR']

#df.to_csv("/filepath/PGA_2025_Data.csv") #<- Uncomment this line to output the final dataset to desired filepath
df
```

```python
#Summary Statistics of relevant columns
df.drop(columns=['RANK', 
                 'PLAYER_ID', 
                 'FAIRWAYS HIT', 
                 'POSSIBLE FAIRWAYS', 
                 'GREENS HIT', 
                 '# HOLES', 
                 'PAR OR BETTER',
                 'MISSED GIR',
                 '# OF PUTTS',
                 '# OF HOLES',
                 'MEASURED ROUNDS',
                 'RELATIVE/PAR',
                 'TOTAL SG:OTT',
                 'TOTAL SG:APP',
                 'TOTAL SG:ARG',
                 'TOTAL SG:PUTTING']).describe()
```
